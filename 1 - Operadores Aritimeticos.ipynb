```python
# Print com Format
nome = input('Digite seu nome: ')
print('Seja bem-vindo, {}!'.format(nome))
#ou
print(f'Seja bem-vindo, {nome}!')
```

```python
# Somando dois números
n1 = float(input('Digite um número:'))
n2 = float(input('Digite outro número:'))
soma = n1+n2
print(f'{n1} + {n2} = {soma}')
```

```python
# Dissecando uma variável
a = input('Digite algo: ')
print('Tipo primitivo: ',type(a))
print('Só tem espaços? ',a.isspace())
print('Só tem números? ',a.isnumeric())
print('Só tem letras? ',a.isalpha())
print('Tem letras e números? ',a.isalnum())
print('Está tudo minusculo? ',a.islower())
print('Está tudo maiusculo?',a.isupper())
print('Só primeira letra maiuscula?',a.istitle())
```

```python
# Antecessor e Sucessor
n = int(input('Digite um número inteiro:'))
print(f'Seu antecessor é {n-1} e seu sucessor é {n+1}')
```

```python
# Dobro, triplo, raiz quadrada
n = float(input('Digite um número:'))
print(f'O dobro de {n} é {n*2}')
print(f'O triplo de {n} é {n*3}')
print(f'A raiz quadrada de {n} é {n**(1/2)}')
```

```python
# Transformação de Unidades
n = float(input('Digite um valor em metros: '))
print(f'{n*0.001}km')
print(f'{n*0.01}hm')
print(f'{n*0.1}dam')
print(f'{n}m')
print(f'{n*10}dm')
print(f'{n*100}cm')
print(f'{n*1000}mm')
```

```python
# Tabuada
n = int(input('Digite um número para ver sua tabuada: '))
c = 0
while c <= 10:
  print(f'{n} * {c} = {n*c}')
  c += 1
```

```python
# Tabuada de 0 a 9 (Comentário)
print('TABUADA DE 0 A 9:')
for a in range(1, 10):
    print('--' *5)
    for b in range(1, 11):
        print(f'{a} x {b:2} = {a*b}')
```

```python
# Conversor de Moedas
real = float(input('Quantos R$ você tem? '))
print(f'Com R$ {real} você pode comprar US$ {(real/5.44):.2f}')
```

```python
# Pintando Parede
l = float(input('Digite a largura da parede em metros: '))
h = float(input('Digite a altura da parede em metros: '))
area = l*h
print(f'Sua parede tem {area:.2f}m² de área.')
# Para cada 2m², 1L de tinta é necessário, portanto:
print(f'Você precisará de {area/2:.2f}l de tinta para pinta-la')
```

```python
# Desconto Percentual
preco = float(input('Qual o preço do produto? R$'))
print(f'Preço do produto com desconto de 5%: R${preco*0.95}')
```

```python
# Desconto Percentual Escolhido Pelo Usuário
preco = float(input('Qual o preço do produto? R$'))
desconto = float(input('Digite o percentual(%) de desconto: '))
print(f'Preço do produto com desconto de 5%: R${preco*(1 - desconto/100):.2f}')
```

```python
# Reajuste Salarial
salario = float(input('Qual o salário do funcionário? R$'))
aumento = float(input('Qual o percentual(%) de aumento: '))/100
print(f'O novo salário do funcionário é: R$ {salario*(1 + aumento):.2f}')
```

```python
# Conversor de Temperaturas
C = float(input('Digite a temperatura em °C: ' ))
F = C*1.8 + 32
K = C + 273.15
print(f'{C}°C é equivalente a {F}°F ou {K}K.')
```

```python
# Aluguel de Carros
D = int(input('Quantidade de dias que o carro foi alugado: '))
km = float(input('Quantidade de km rodados: '))
preco = D*60 + km*0.15
print(f'Preço a pagar: R${preco:.2f}')
```
